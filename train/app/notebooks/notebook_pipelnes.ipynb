```python
# Imports

# pandas
import pandas as pd
from pandas import Series,DataFrame

# numpy, matplotlib, seaborn
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline

# machine learning
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, LinearSVC
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
```

```python
# vamos a hacer nuestro propio split para tener la Y en el test set y poder obtener el score sobre el test set, ademas de poder comprobar si nos estaria haciendo overfitting por ejemplo
df    = pd.read_csv("../data/data.csv")
titanic_df, test_df = train_test_split(df, test_size=0.2, random_state=50,stratify=df['Survived'])
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.pipeline import Pipeline



class titanicPipeline(BaseEstimator, TransformerMixin):
    '''
        all feature engineer in one transformer
    '''
    def __init__(self, columns=None):

        self.columns = columns
    

    def fit(self, X, y=None):

        return self

    def transform(self, X, y=None):

        # dont modify original make a copy
        _X = X.copy()

        # Cabin
        # It has a lot of NaN values, so it won't cause a remarkable impact on prediction in this simple model
        _X.drop("Cabin",axis=1,inplace=True)
        
        #drop un unnecesary cols
        _X.drop(self.columns, axis=1, inplace=True)

        #embarked
        _X = self.embarkedStep(_X)

        #fare
        _X = self.fareStep(_X)

        #age
        _X = self.ageStep(_X)

        #sex
        _X = self.sexStep(_X)

        # return
        return _X    

    def embarkedStep(self, X):
        # only in titanic_df, fill the two missing values with the most occurred value, which is "S".
        X["Embarked"] = X["Embarked"].fillna("S")

        # Either to consider Embarked column in predictions,
        # and remove "S" dummy variable, 
        # and leave "C" & "Q", since they seem to have a good rate for Survival.

        # OR, don't create dummy variables for Embarked column, just drop it, 
        # because logically, Embarked doesn't seem to be useful in prediction.

        embark_dummies_titanic  = pd.get_dummies(X['Embarked'])
        embark_dummies_titanic.drop(['S'], axis=1, inplace=True)

        X = X.join(embark_dummies_titanic)
        X.drop(['Embarked'], axis=1,inplace=True)

        return X

    def fareStep(self, X):
        
        # convert from float to int
        X['Fare'] = X['Fare'].astype(int)

        return X


    def ageStep(self, X):
        # get average, std, and number of NaN values in titanic_df
        average_age_titanic   = X["Age"].mean()
        std_age_titanic       = X["Age"].std()
        count_nan_age_titanic = X["Age"].isnull().sum()

        # generate random numbers between (mean - std) & (mean + std)
        rand_v = np.random.randint(average_age_titanic - std_age_titanic, average_age_titanic + std_age_titanic, size = count_nan_age_titanic)

        # drop all null values, and convert to int
        X['Age'].dropna().astype(int)

        # fill NaN values in Age column with random values generated
        X["Age"][np.isnan(X["Age"])] = rand_v

        # convert from float to int
        X['Age'] = X['Age'].astype(int)

        return X

    def sexStep(self, X):

        X['Person'] = X[['Age','Sex']].apply(self.get_person,axis=1)

        # No need to use Sex column since we created Person column
        X.drop(['Sex'],axis=1,inplace=True)

        # create dummy variables for Person column, & drop Male as it has the lowest average of survived passengers
        person_dummies_titanic  = pd.get_dummies(X['Person'])
        person_dummies_titanic.columns = ['Child','Female','Male']
        person_dummies_titanic.drop(['Male'], axis=1, inplace=True)

        X = X.join(person_dummies_titanic)

        X.drop(['Person'],axis=1,inplace=True)

        return X


    def get_person(self, passenger):
        age,sex = passenger
        return 'child' if age < 16 else sex
```

```python
# define training and testing sets

X_train = titanic_df.drop("Survived",axis=1)
Y_train = titanic_df["Survived"]

X_test  = test_df.drop(["Survived"],axis=1).copy()
Y_test = test_df["Survived"]
```

```python
# Logistic Regression with pipeline
logregPipeline = Pipeline(
    steps=[
        ("data_transform",titanicPipeline(columns  = ['PassengerId','Name','Ticket'])),
        ("model",LogisticRegression())
    ]
)

#parameters
parameters = [{'model__solver': ['newton-cg', 'lbfgs','sag', 'saga'],
                'model__C':[0.5,1,1.5,2,2.5],
                'model__class_weight':[None,'balanced'],
                'model__max_iter':[1000,2000,3000],
                'model__penalty':['l1','l2']}]


logistic_version = GridSearchCV(logregPipeline, parameters, n_jobs=-1)
logistic_version.fit(X_train, Y_train)
logreg_score = logistic_version.score(X_test,Y_test)

print('Training set score: ' + str(logistic_version.score(X_train,Y_train)))
print('Test set score: ' + str(logreg_score))
print('GridSearchCV params: ', logistic_version.best_params_)
```

```python
#Support Vector Machines
svcPipeline = Pipeline(
    steps=[
        ("data_transform",titanicPipeline(['PassengerId','Name','Ticket'])),
        ("model",SVC())
    ]
)
#cv params
parameters = {'model__C': [0.1,1, 10, 100], 'model__gamma': [1,0.1,0.01,0.001]}

svm_version = GridSearchCV(svcPipeline, parameters, n_jobs=-1)
svm_version.fit(X_train, Y_train)
svm_score = svm_version.score(X_test,Y_test)

print('Training set score: ' + str(svm_version.score(X_train,Y_train)))
print('Test set score: ' + str(svm_score))
print('GridSearchCV params: ', svm_version.best_params_)
```

```python
# Random Forests
rfPipeline = Pipeline(
    steps=[
        ("data_transform",titanicPipeline(['PassengerId','Name','Ticket'])),
        ("model",RandomForestClassifier())
    ]
)
#cv params
parameters = { 
    'model__n_estimators': [200, 500],
    'model__max_features': ['auto', 'sqrt', 'log2'],
    'model__max_depth' : [4,5,6,7,8],
    'model__criterion' :['gini', 'entropy']
}

rf_version = GridSearchCV(rfPipeline, parameters, n_jobs=-1)
rf_version.fit(X_train, Y_train)
random_forest_score = rf_version.score(X_test,Y_test)

print('Training set score: ' + str(rf_version.score(X_train,Y_train)))
print('Test set score: ' + str(random_forest_score))
print('GridSearchCV params: ', rf_version.best_params_)
```

```python
# Gradient Boosts
gbPipeline = Pipeline(
    steps=[
        ("data_transform",titanicPipeline(['PassengerId','Name','Ticket'])),
        ("model",GradientBoostingClassifier())
    ]
)
#cv params
parameters = {
    "model__max_depth":[3,5,8],
    "model__max_features":["log2","sqrt"]
    }

gb_version = GridSearchCV(gbPipeline, parameters, n_jobs=-1)
gb_version.fit(X_train, Y_train)
gb_score = gb_version.score(X_test,Y_test)

print('Training set score: ' + str(gb_version.score(X_train,Y_train)))
print('Test set score: ' + str(gb_score))
print('GridSearchCV params: ', gb_version.best_params_)
```

```python
#lets compare models
modelos=[
    {
        'name':'Logistic Regression',
        'score':logreg_score
    },
    {
        'name':'Support Vector Machines',
        'score':svm_score
    },
    {
        'name':'Random Forests',
        'score':random_forest_score
    },
    {
        'name':'Gradient Boosts',
        'score':gb_score
    }
]

modelos_df = DataFrame(modelos)
modelos_df.sort_values(by='score', ascending=False)
```
